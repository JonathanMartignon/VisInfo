```python
from os import listdir
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def muestra_dirs(ruta):
    '''
    La presente función me ayudaba a navegar entre los archivos sin 
    la necesidad de salir de Jupyter. Sólo había que establecer la 
    ruta a buscar.
    '''
    print("Ruta actual: ",ruta,'\n')
    for i,j in enumerate(listdir(ruta)):
        print(f"{i+1}.- {j}")
```

```python
Felicidad_Mex = pd.read_csv('./DatosMexico/Ranking de Felicidad Mexico.csv')
Felicidad_Fin = pd.read_csv('./Comparaciones/Felicidad/Ranking de Felicidad Finlandia.csv')
Felicidad_Nor = pd.read_csv('./Comparaciones/Felicidad/Ranking de Felicidad Noruega.csv')
Felicidad_Sui = pd.read_csv('./Comparaciones/Desarrollo/Ranking Índice de Felicidad Suiza.csv')
```

```python
miny = min(Felicidad_Mex['Ranking'].min(),Felicidad_Fin['Ranking'].min(),
           Felicidad_Nor['Ranking'].min(),Felicidad_Sui['Ranking'].min())

maxy = max(Felicidad_Mex['Ranking'].max(),Felicidad_Fin['Ranking'].max(),
           Felicidad_Nor['Ranking'].max(),Felicidad_Sui['Ranking'].max())

fig = plt.figure(figsize=(14,6))

plt.title('Comparación Ranking de Felicidad')
plt.grid()
plt.plot(Felicidad_Mex['Año'],Felicidad_Mex['Ranking'],'>-',color = 'forestgreen',label='México')
plt.plot(Felicidad_Fin['Año'],Felicidad_Fin['Ranking'],'>-',color = 'darkblue',label='Finlandia')
plt.plot(Felicidad_Nor['Año'],Felicidad_Nor['Ranking'],'>-',color = 'olive',label='Noruega')
plt.plot(Felicidad_Sui['Año'],Felicidad_Sui['Ranking'],'>-',color = 'red',label='Suiza')
plt.legend()
# plt.xticks(np.arange(minx,maxx+1))
# plt.xlim(minx-1,maxx+1)
plt.yticks(np.arange(miny,maxy+4,2))
plt.ylim(miny-1,maxy+2)
plt.ylabel('Posición')
plt.xlabel('Año')


plt.show()
```

```python
Brecha_gen_Mex = pd.read_csv('./DatosMexico/Brecha de Género México.csv')
Brecha_gen_Fin = pd.read_csv('./Comparaciones/Felicidad/Brecha Género Finlandia.csv')
Brecha_gen_Nor = pd.read_csv('./Comparaciones/Felicidad/Brecha Género Noruega.csv')
Brecha_gen_Sui = pd.read_csv('./Comparaciones/Desarrollo/Brecha de Género Suiza.csv')
```

```python
miny = min(Brecha_gen_Mex['Índice de Brecha de Género'].min(),
           Brecha_gen_Fin['Índice de Brecha de Género'].min(),
           Brecha_gen_Nor['Índice de Brecha de Género'].min(),
           Brecha_gen_Sui['Índice de Brecha de Género'].min())

maxy = max(Brecha_gen_Mex['Índice de Brecha de Género'].max(),
           Brecha_gen_Fin['Índice de Brecha de Género'].max(),
           Brecha_gen_Nor['Índice de Brecha de Género'].max(),
           Brecha_gen_Sui['Índice de Brecha de Género'].max())

fig = plt.figure(figsize=(14,6))

plt.title('Comparación Índice de Brecha de género')
plt.grid()

plt.plot(Brecha_gen_Mex['Año'],Brecha_gen_Mex['Índice de Brecha de Género'],
         'o-',color = 'forestgreen',label='México')

plt.plot(Brecha_gen_Fin['Año'],Brecha_gen_Fin['Índice de Brecha de Género'],
         '>-',color = 'darkblue',label='Finlandia')

plt.plot(Brecha_gen_Nor['Año'],Brecha_gen_Nor['Índice de Brecha de Género'],
         '>-',color = 'olive',label='Noruega')

plt.plot(Brecha_gen_Sui['Año'],Brecha_gen_Sui['Índice de Brecha de Género'],
         '>-',color = 'red',label='Suiza')

plt.legend()

plt.yticks(np.arange(miny,maxy+0.1,0.05))
plt.ylim(miny-0.05,maxy+0.05)
plt.ylabel('Posición')
plt.xlabel('Año')


plt.show()
```

```python
Deuda_Mex = pd.read_csv('./DatosMexico/Deuda Mexico.csv')[:-4]
Deuda_Fin = pd.read_csv('./Comparaciones/Felicidad/Deuda Finlandia.csv')[1:-4]
Deuda_Fin.reset_index(inplace=True)
Deuda_Nor = pd.read_csv('./Comparaciones/Felicidad/Deuda Noruega.csv')[:-4]
Deuda_Sui = pd.read_csv('./Comparaciones/Desarrollo/Deuda Suiza.csv')[:-4]
```

```python
fig = plt.figure(figsize=(12,15))
nombres = ["México","Finlandia","Noruega","Suiza"]

plt.suptitle('Comparación de la Deuda pública por país')



for i in range(Deuda_Mex.shape[0]):
    
    plt.subplot(4,4,i+1)
    
    plt.title(Deuda_Mex['Año'][i])
    
    if i == 0:
        
        Valores = [Deuda_Mex['Deuda en Millones de Pesos MXN'][i],
                   Deuda_Fin['Deuda en Millones de Pesos MXN'][i],
                   Deuda_Nor['Deuda en Millones de Pesos MXN'][i]]
    else:
        Valores = [Deuda_Mex['Deuda en Millones de Pesos MXN'][i],
                   Deuda_Fin['Deuda en Millones de Pesos MXN'][i],
                   Deuda_Nor['Deuda en Millones de Pesos MXN'][i],
                   Deuda_Sui['Deuda en Millones de Pesos MXN'][i-1]
                  ]
    
    plt.pie(Valores,autopct='%0.2f%%',colors=['limegreen','darkblue','mediumaquamarine','Red'],pctdistance=1.21)
    plt.legend(nombres,loc='best')
fig.tight_layout()
plt.show()
```
