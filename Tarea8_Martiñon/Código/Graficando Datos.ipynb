```python
import matplotlib.pyplot as plt
from os import listdir
import pandas as pd
import numpy as np
```

# Extrayendo la información sobre México de nuestros CSV

```python
# Guardamos en una lista nuestros datasets a utilizar
# Esto es porque si leíamos datos del directorio podían
# Variar en cuanto al orden y eso ocasionaría un caos.
# Mejor declaramos directamente el orden, pues conocemos 
# La distribución de los datos.

names = ['Homicidios intencionados México',
 'Brecha de Género México',
 'Ranking de Felicidad Mexico',
 'Deuda Mexico',
 'Depresion en el hogar en Mexico',
 'Indice de desarrollo humano México',
 'Indice Global de Paz México',
 'Suicidios México',
 'Religion en México',
 'Defunciones en Mexico',
 'Reclusos en Mexico',
 'Matrimonios en México']

#Mostrando el número de nombres de datos
len(names)
```

```python
data_Mex = []

# A partir de los nombres, leeremos los datasets
# Y los agregaremos a una lista en el mismo orden
# Que los nombres, así obteniendo la posición del
# Nombre, conocemos la posición del dataset
for data in names:
    
    data_Mex.append(pd.read_csv('./DatosMexico/'+data+'.csv'))

#Mostrando la cantidad de datos recolectados
#Para verificar que sean los mismos que nombres.
len(data_Mex)
```

# Graficando los Datos de México

Una vez que leímos los datos y estamos seguros de su ubicación, podemos realizar los gráficos de cada Conjunto

```python
#Mostraremos los rankings en los que aparece México

minimo = min(min(data_Mex[1]['Año']),min(data_Mex[2]['Año']),min(data_Mex[5]['Año']),min(data_Mex[6]['Año']))
maximo = max(max(data_Mex[1]['Año']),max(data_Mex[2]['Año']),max(data_Mex[5]['Año']),max(data_Mex[6]['Año']))
fig = plt.figure(figsize=(14,6))

plt.title("Rankings de México a nivel mundial de acuerdo con datosmacro.expansion.com")
plt.grid()
plt.plot(data_Mex[1]['Año'],data_Mex[1]['Ranking'],'>-',color='mediumorchid',label = names[1])
plt.plot(data_Mex[2]['Año'],data_Mex[2]['Ranking'],'o-',color='chartreuse',label = names[2])
plt.plot(data_Mex[5]['Año'],data_Mex[5]['Ranking'],'v-',color='deepskyblue',label = names[5])
plt.plot(data_Mex[6]['Año'],data_Mex[6]['Ranking'],'s-',color='darkorange',label = names[6])

plt.xlabel('Año')
plt.xticks(np.arange(minimo,maximo+1))
plt.xlim(minimo-1,maximo+1)
plt.ylabel('Posición')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))

plt.subplot(2,1,1)

plt.title('Población Cristiana en México')
plt.xlabel('Año')
plt.xticks(data_Mex[8]['Año'])
plt.ylabel('Porcentaje')
plt.grid()
plt.plot(data_Mex[8]['Año'],data_Mex[8]['Cristianismo %'],'*-',color='red')
plt.yticks(np.arange(93,98.5,0.5))
plt.ylim(93,98)

plt.subplot(2,1,2)

plt.title('Población Judía en México')
plt.xlabel('Año')
plt.xticks(data_Mex[8]['Año'])
plt.ylabel('Porcentaje')
plt.grid()
plt.plot(data_Mex[8]['Año'][:-4],data_Mex[8]['Judaísmo %'][:-4],'o-',color ='midnightblue')
plt.text(1978,0.06,"Falta dato",fontsize=12,color='Red')
plt.plot(data_Mex[8]['Año'][-3:],data_Mex[8]['Judaísmo %'][-3:],'o-', color = 'midnightblue')
plt.ylim(0.02,0.14)


plt.tight_layout()

plt.show()
```
